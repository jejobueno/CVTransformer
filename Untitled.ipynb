```python
from tika import parser # pip install tika
import json
import spacy
import re

from utils.Solitan import Solitan

pdf_file = r'assets/Vladimir_Anaimanovich_2022-06-02T12_28_56.954Z.pdf'

cv = parser.from_file(pdf_file)['content']
cv = re.sub('\n+', '\n', cv)
cv = re.sub('-\n+', '', cv)
cv = re.sub("^[a-zA-Z0-9]*$", '', cv)
print(cv)
```

```python

cv_in_lines = cv.splitlines()
while("" in cv_in_lines):
    cv_in_lines.remove("")

solitan = Solitan()

nlp = spacy.load("en_core_web_lg")
nlp.get_pipe("ner").labels

print(solitan)
print(cv)

line_with_name = 0

for i in range(0, len(cv_in_lines)):
    doc = nlp(cv_in_lines[i])
    for ent in doc.ents:
        if ent.label_ == "PERSON" and solitan.name == "":
            solitan.name = ent.text
            solitan.rol = cv_in_lines[i+1]
            break
    print(ent.label_, ent.text)

print(solitan.name)
```

```python
sections = ['Personal Info', 'Strengths', 'Work history', 'Education', 'Certificates', 'Projects', 'Skills', 'Web presence', 'Languages', 'Hobbies & passions']
```

## Process Work Experience

```python
from spacy.matcher import Matcher

matcher = Matcher(nlp.vocab)
pattern = [{"TEXT": {"REGEX": '[0-9]{2}/[0-9]{4}'}}]
work = cv_in_sections['Work history']
```

```python
re.findall('([0-9]{2}/[0-9]{4}—( |[0-9]{2}/[0-9]{4}))', work)
```

```python
matcher.add('Date',[pattern])

doc = nlp(work)
date_matches = matcher(doc)
for i in range(0,len(date_matches)):
    match_id, start, end = date_matches[i]
    string_id = nlp.vocab.strings[match_id]  # Get string representation
    span_date = doc[start:end]  # The matched span
    if i < len(date_matches) - 1:
        span_jobdescription = doc[end:date_matches[i+1][1]].text.splitlines()
    else:
        span_jobdescription = doc[end::].text.splitlines()
    span_jobtitle, span_company = span_jobdescription.pop(0).strip("— .").split(',')
    span_jobdescription = " ".join(span_jobdescription)
    print(match_id, string_id, start, end, span_date.text)
    print('#################')
    print(span_jobtitle)
    print(span_jobdescription)
```

## Process Education

```python
matcher = Matcher(nlp.vocab)
matcher_lower = Matcher(nlp.vocab)

pattern = [{"TEXT": {"REGEX": '[0-9]{4}'}}]
pattern_lower = [{"TEXT": {"REGEX": '[A-Z]*[a-z]+'}}]

work = cv_in_sections['Education']

matcher.add('Date',[pattern])
matcher_lower.add('Lower case text',[pattern_lower])


doc = nlp(work)
date_matches = matcher(doc)
for i in range(0,len(date_matches)):
    match_id, start, end = date_matches[i]
    string_id = nlp.vocab.strings[match_id]  # Get string representation
    span_date = doc[start:end].text  # The matched span
    if i < len(date_matches) - 1:
        span_jobdescription = re.sub('\n', ' ', doc[end:date_matches[i+1][1]].text)
    else:
        span_jobdescription = re.sub('\n', ' ', doc[end::].text)
        
    education_doc = nlp(span_jobdescription)
    lower_matches = matcher_lower(education_doc)
    span_jobtitle = education_doc[:lower_matches[0][1]].text
    span_jobdescription = education_doc[lower_matches[0][1]::].text
    span_jobtitle, span_company = span_jobtitle.strip("— .").split(',')
    print('#################')
    print(span_date)
    print(span_jobtitle)
    print(span_company.strip("— ."))
    print(span_jobdescription)
```
